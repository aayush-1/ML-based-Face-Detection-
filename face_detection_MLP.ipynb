```python
#neural network with sgd solver





from scipy import misc
import numpy as np
from sklearn.decomposition import PCA
from sklearn.neural_network import MLPClassifier
from sklearn import metrics

from sklearn.model_selection import train_test_split
import os
import matplotlib.pyplot as plt
path1="data_set1"
imagepaths1=[os.path.join(path1,f) for f in os.listdir(path1) if f.endswith('.jpg')]
#print(imagepaths1)
b=[]
l=1

for imagepath1 in imagepaths1:
    a=misc.imread(imagepath1)
    #print(a.shape)
    b=a.reshape(1,196*196)
    
    if(l==1):
        positive=b
        l=0
    else:
        positive=np.vstack((positive,b))
        
print (positive.shape)

path0="data_set0"
imagepaths0=[os.path.join(path0,f) for f in os.listdir(path0) if f.endswith('.jpg')]
#print(len(imagepaths0))

l=1

for imagepath0 in imagepaths0:
    a=misc.imread(imagepath0)
    #print(a.shape)
    b=a.reshape(1,196*196)
    
    if(l==1):
        negative=b
        l=0
    else:
        negative=np.vstack((negative,b))

print(negative.shape)   


label1=np.ones((50,1))
label0=np.zeros((50,1))
nd1=np.hstack((positive,label1))

nd0=np.hstack((negative,label0))

newdata=np.vstack((nd1,nd0))
a=np.random.randint(0,100, size=100)
newdata=newdata[a,:]
print (newdata.shape)

x1=newdata[0:20,:38416]
y1=newdata[0:20,38416]
x2=newdata[20:40,:38416]
y2=newdata[20:40,38416]
x3=newdata[40:60,:38416]
y3=newdata[40:60,38416]
x4=newdata[60:80,:38416]
y4=newdata[60:80,38416]
x5=newdata[80:100,:38416]
y5=newdata[80:100,38416]



data=[x1,x2,x3,x4,x5]
label=[y1,y2,y3,y4,y5]

acc=0
#10cv  

for j in range(0,5):
    if(j==0):
        xtrain=data[1]
        l=2
        while(l<5):
            xtrain=np.vstack((xtrain,data[l]))
            l=l+1
    else:
        xtrain=data[0]
        l=1
        while(l<5):
            if(l==j):
                l=l+1
            else:
                xtrain=np.vstack((xtrain,data[l]))
                l=l+1
    if(j==0):
        labeltrain=label[1]
        l=2
        while(l<5):
            labeltrain=np.hstack((labeltrain,label[l]))
            l=l+1
    else:
        labeltrain=label[0]
        l=1
        while(l<5):
            if(l==j):
                l=l+1
            else:
                labeltrain=np.hstack((labeltrain,label[l]))
                l=l+1
    

    xtest=data[j]
    labeltest=label[j]
    
    #pca for train data
    pca=PCA(n_components=20).fit(xtrain)
    xtrainpca=pca.transform(xtrain)
    print("xtrainpca=",xtrainpca.shape)
    
    #pca for test data
    pc=PCA(n_components=20).fit(xtest)
    xtestpca=pc.transform(xtest)
    print("xtestpca=",xtestpca.shape)
    
    
    
    

    






#forward feature selection

    accffs=[]
    clf = MLPClassifier(solver='sgd', max_iter=2000 , alpha=0.1,hidden_layer_sizes=(5,5), random_state=1)
    for i in range(0,20):
        
        xtr,xt,ytrain,ytest=train_test_split(xtrainpca[:,:(i+1)],labeltrain,test_size=0.25,random_state=None)
        clf.fit(xtr,ytrain)
        ypredicted=clf.predict(xt)
        accffs.append(metrics.accuracy_score(ytest, ypredicted))
        print(accffs[i])
    

    print(accffs)



    a=np.linspace(0,20,20)
    plt.plot(a,accffs)
    plt.show()
    
    
    b=int(input("enter the no. features to take for further testing"))
    
    xfinaltrain=xtrainpca[:,:b]
    xfinaltest=xtestpca[:,:b]
    clf.fit(xfinaltrain,labeltrain)
    ypredicted=clf.predict(xfinaltest)
    accuracy=metrics.accuracy_score(labeltest, ypredicted)
    acc=acc + accuracy

    print("\n\naccuracy: ",accuracy)
    
print("\n\nfinal and avg accuracy for 5 cv is : ",acc/5)
    



```
