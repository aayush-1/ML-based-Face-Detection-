```python
#photo for final show


import cv2
import matplotlib.pyplot as plt
import numpy as np
faceDetect=cv2.CascadeClassifier('C:\\Users\\Aayush Arora\\Anaconda3\\pkgs\\opencv-3.3.1-py36h20b85fd_1\\Library\\etc\\haarcascades\\haarcascade_frontalface_default.xml')
cam=cv2.VideoCapture(0)
id=input("enter name:\t")
no=0
while(cam.isOpened()):
    ret,img=cam.read()
    if ret is True:
        
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    else:
        continue
    faces=faceDetect.detectMultiScale(gray,1.3,5)
    for (x,y,w,h) in faces:
       
        if(w>196 and h>196):
            no=no+1
            cv2.putText(img,str(w )+str(h),(x,y+h),cv2.FONT_HERSHEY_SIMPLEX,2,(255,255,255))
            cv2.imwrite("C:\\Users\\Aayush Arora\\testingdata1\\" + str(id) + "." + str(no) + ".jpg",gray[y:y+h,x:x+w])
            
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),2)
        cv2.waitKey(100)
    cv2.imshow("face",img)
    cv2.waitKey(1)
        
        
    
    
    if(no>49):
        break
cam.release()
cv2.destroyAllWindows()  


faceDetect=cv2.CascadeClassifier('C:\\Users\\Aayush Arora\\Anaconda3\\pkgs\\opencv-3.3.1-py36h20b85fd_1\\Library\\etc\\haarcascades\\haarcascade_frontalface_default.xml')
cam=cv2.VideoCapture(0)

 
```

```python
#no cross validation ....just the simple code by me

from scipy import misc
import numpy as np
from sklearn.decomposition import PCA
from sklearn import metrics
from sklearn import svm
from sklearn.model_selection import train_test_split
import os
import matplotlib.pyplot as plt
path1="data_set1"
imagepaths1=[os.path.join(path1,f) for f in os.listdir(path1) if f.endswith('.jpg')]
#print(imagepaths1)
b=[]
l=1

for imagepath1 in imagepaths1:
    a=misc.imread(imagepath1)
    #print(a.shape)
    a=misc.imresize(a,[200,200])
    b=a.reshape(1,200*200)
    
    if(l==1):
        positive=b
        l=0
    else:
        positive=np.vstack((positive,b))
        
print (positive.shape)

path0="data_set0"
imagepaths0=[os.path.join(path0,f) for f in os.listdir(path0) if f.endswith('.jpg')]
#print(len(imagepaths0))

l=1

for imagepath0 in imagepaths0:
    a=misc.imread(imagepath0)
    #print(a.shape)
    a=misc.imresize(a,[200,200])
    b=a.reshape(1,200*200)
    
    if(l==1):
        negative=b
        l=0
    else:
        negative=np.vstack((negative,b))

print(negative.shape)   


label1=np.ones((50,1))
label0=np.zeros((50,1))
nd1=np.hstack((positive,label1))

nd0=np.hstack((negative,label0))

newdata=np.vstack((nd1,nd0))
a=np.random.randint(0,100, size=100)
newdata=newdata[a,:]
x=newdata[0:100,:40000]
y1=newdata[0:50,40000]

y2=newdata[50:100,40000]

print(newdata.shape)

pca=PCA(n_components=50).fit(x)
xtrainpca=pca.transform(x)
x1=xtrainpca[0:50,:]
x2=xtrainpca[50:100,:]
accffs=[]


    
clf=svm.SVC(kernel='linear')
for i in range(0,50):
        
    xtr,xt,ytrain,ytest=train_test_split(x1[:,:(i+1)],y1,test_size=0.25,random_state=None)
    clf.fit(xtr,ytrain)
    ypredicted=clf.predict(xt)
    accffs.append(metrics.accuracy_score(ytest, ypredicted))
    print(accffs[i])
    

print(accffs)



a=np.linspace(0,50,50)
plt.plot(a,accffs)
plt.show()
    
    
b=int(input("enter the no. features to take for further testing"))  
xfinaltrain=x1[:,:b]
xfinaltest=x2[:,:b]
clf.fit(xfinaltrain,y1)
ypredicted=clf.predict(xfinaltest)
accuracy=metrics.accuracy_score(y2, ypredicted)

print("\n\naccuracy: ", accuracy)
    




```
